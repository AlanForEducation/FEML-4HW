```python
import pandas as pd
```

```python
data = pd.read_csv('adult_csv.csv')
data.head()
```

```python
data.info()
```

```python
data = data.dropna() #удаление пустых значений
data.info()
```

```python
# колонки для dummies
col_for_dummies = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country'] 

# колонки для обучения 
selected_cols = data[['age', 'workclass', 'fnlwgt', 'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capitalgain', 'capitalloss', 'hoursperweek', 'native-country']]
```

```python
X = pd.get_dummies(selected_cols, columns=col_for_dummies)
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(data['class'])
```

```python
le.classes_
```

```python
y = le.fit_transform(data['class'])
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

## Логистическа регрессия

```python
model_scaler = make_pipeline(StandardScaler(), LogisticRegression())
model = LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model.fit(X_train, y_train)
model_scaler.fit(X_train, y_train)
```

```python
print(f'score без StandardScaler = {model.score(X_test, y_test)}')
print(f'score c StandardScaler = {model_scaler.score(X_test, y_test)}')
```

## SVM

```python
from sklearn.svm import SVC
```

```python
clf_scaler = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf = SVC(gamma='auto')
```

```python
clf_scaler.fit(X_train, y_train)
```

```python
clf.fit(X_train, y_train)
```

```python
print(f'score без StandardScaler = {clf.score(X_test, y_test)}')
print(f'score c StandardScaler = {clf_scaler.score(X_test, y_test)}')
```
